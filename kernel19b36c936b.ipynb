# converting the images into array and assigning each input its respective label

##Here we are using max 100 from each class

```python
train_images=[]
labels=[]
for i in vehicle_list:
    img_dir=os.path.join(train,i)
    feature_imgs=os.listdir(img_dir)
    #print(feature_imgs)
    #break
    count=0
    for j in feature_imgs:
        if count==100:
            break
        img=os.path.join(img_dir,j)
        img = image.load_img(img, target_size=(img_size, img_size))
        img = image.img_to_array(img)
        img = img/255
        train_images.append(img)
        labels.append(vehicle_list.index(i))
        count+=1
       
```

# shuffle the dataset

```python
temp = list(zip(train_images, labels)) 
random.shuffle(temp) 
train_images, labels = zip(*temp)
```

```python
X = np.array(train_images)
```

# splitting

```python

X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.1, random_state=101)
print(X_train.shape)
print(X_test.shape)
```

```python

y_train_labels = to_categorical(y_train)
print(y_train_labels.shape)
```

# Model Network

```python
def build(width, height, depth, classes):
    #initialize the model along with the input shape
    model = Sequential()
    inputShape = (height, width, depth)
    chanDim = -1
    
    if K.image_data_format() == 'channels_first':
        inputShape = (depth, height, width)
        chanDim = 1
        
    # CONV -> RELU -> MAXPOOL
    model.add(Convolution2D(64, (3,3), padding='same', input_shape=inputShape))
    model.add(Activation('relu'))
    model.add(BatchNormalization(axis=chanDim))
    model.add(MaxPooling2D(pool_size=(3,3)))
    model.add(Dropout(0.25))
    
    # (CONV -> RELU)*2 -> AVGPOOL
    model.add(Convolution2D(128, (3,3), padding='same'))
    model.add(Activation('relu'))
    model.add(BatchNormalization(axis=chanDim))
    model.add(MaxPooling2D(pool_size=(3,3)))
    model.add(Dropout(0.25))

    
   
    # DENSE -> RELU
    model.add(Flatten())
    model.add(Dense(128))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))
    
    # DENSE -> RELU
    model.add(Dense(64))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))
    
    # sigmoid -> just to check the accuracy with this (softmax would work too)
    model.add(Dense(classes))
    model.add(Activation('softmax'))
    
    return model
```

```python

model = build(img_size, img_size, 3, 17)
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['acc'])
model.summary()
```

# Training Model

```python
history = model.fit(X_train, y_train_labels, batch_size=16, epochs=120, validation_split=0.2)
```

# Plotting the results

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
